In Python, an anonymous function is a function that is defined without a name

Anonymous functions are defined using a lambda keyword. Therefore it is also called lambda function

**Syntex of lambda function**<br>
lambda arguments:expression<br>

lambda functions can have any no. of arguments but only one expression.

```python
insurance_details.groupby(by='region',axis=0)['charges'].mean().sort_values(ascending=False)
```

#### 2. Get the average insurance charges based on smokers and region 

```python
insurance_details.groupby(by=['region','smoker'])['charges'].mean().sort_values(ascending=False)
```
