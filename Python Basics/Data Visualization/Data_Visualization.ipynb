### 1. Univariate Analysis

To understand the quality of one feature.

```python
import matplotlib.pyplot as plt
%matplotlib inline
#from matplotlib import pyplot as plt
```

### 1.1 Histogram

```python
#Histogram
plt.figure(figsize=(6,4))
plt.hist(x='mpg',data=mtcars_data)
plt.xlabel('MPG')
plt.ylabel('Frequancy')
plt.title('MPG Distribution',size=15)# * Variable length arg, ** key word argument 
plt.show()
```

**Histogram** is a univariate analysis and always we have to give histogram a continuous data as it says distribution of the data.

#### 1.2 Count Plot

```python
mtcars_data.head()
```

```python
import seaborn as sns
```

```python
mtcars_data['carb'].value_counts()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='carb',y=None,data=mtcars_data)
plt.title('Carborators Distribution',size=16)
plt.show()
```

**Count plot** is used to know the distribution of descrete data.

#### 1.3 Pie Chart

```python
mtcars_data.head()
```

```python
mtcars_data['gear'].value_counts()
```

```python
plt.figure(figsize=(8,6))
plt.pie(x=mtcars_data['gear'].value_counts(), labels=[3,4,5], autopct='%1.2f%%',explode=[0.02,0.02,0.02])
plt.show()
```

### 2. Bi-variate Analysis

To understand relationship between two features

```python
mtcars_data.head()
```

#### 2.1 Scatter plot

```python
plt.scatter(x='hp',y='mpg',data=mtcars_data)
plt.xlabel('HP')
plt.ylabel('MPG')
plt.title('HP vs MPG')
plt.show()
```

```python
plt.scatter(x='wt',y='mpg',data=mtcars_data)
plt.xlabel('WT')
plt.ylabel('MPG')
plt.title('WT vs MPG')
plt.show()
```

**Scatter plot** is used to understand the quality/ relation between two features and two features should be continuous data

#### 2.2 Bar plot

```python
mtcars_data.head()
```

```python
mtcars_data.groupby('cyl')['mpg'].mean()
```

```python
sns.barplot(x='cyl',y='mpg',data=mtcars_data)
plt.show()
```

**Bar plot** takes descrete value in x axis and continuous value in y axis

### 3. Multivariate Analysis

To know relation between multiple features.

#### 3.1 Pair plot

```python
mtcars_data.head()
```

```python
sns.pairplot(data=mtcars_data)
```

Correlation tells us how strong the relation and in which direction the relation is there b/w two variables
Correlation lies b/w -1 to 1

```python
mtcars_corr_matrix=mtcars_data.corr().round(2)
mtcars_corr_matrix
```

#### 3.2 Heat Map

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=mtcars_corr_matrix,annot=True)
```
